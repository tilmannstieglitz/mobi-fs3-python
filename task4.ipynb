**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] < 1 or p[1] < 1 or p[0] >= img_in.shape[0] - 1 or p[1] >= img_in.shape[1] - 1:
            result[p] = 0
        else:
            neighborhood_sum = neighborhood_sum + sum(img_in[(p[0]-1):(p[0]+1), p[1]-1]) - sum(img_in[(p[0]-1):(p[0]+1), p[1]+1])
            result[p] = neighborhood_sum/(6)       

    return result
    
def prewitt_v(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] < 1 or p[1] < 1 or p[0] >= img_in.shape[0] - 1 or p[1] >= img_in.shape[1] - 1:
            result[p] = 0
        else:
            neighborhood_sum = neighborhood_sum + sum(img_in[p[0]-1, (p[1]-1):p[1]+1]) - sum(img_in[p[0]+1, (p[1]-1):p[1]+1])
            result[p] = neighborhood_sum/(6)        

    return result
```

## Sobel filters (bonus)

**<a id="#">Task 4.3.1.</a>** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical 3&times;3 Sobel filtering, respectively:

```python
def sobel_h(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] < 1 or p[1] < 1 or p[0] >= img_in.shape[0] - 1 or p[1] >= img_in.shape[1] - 1:
            result[p] = 0
        else:
            neighborhood_sum = neighborhood_sum + img_in[(p[0]-1), p[1]-1] + 2*img_in[p[0], p[1]-1] + img_in[(p[0]+1), p[1]-1] - (img_in[(p[0]-1), p[1]+1] + 2*img_in[p[0], p[1]+1] + img_in[(p[0]+1), p[1]+1])
            result[p] = neighborhood_sum/(6)       

    return result
    
def sobel_v(img_in):
    result = numpy.zeros(img_in.shape)
    for p in numpy.ndindex(img_in.shape):
        neighborhood_sum = 0
        if p[0] < 1 or p[1] < 1 or p[0] >= img_in.shape[0] - 1 or p[1] >= img_in.shape[1] - 1:
            result[p] = 0
        else:
            neighborhood_sum = neighborhood_sum + img_in[(p[0]-1), p[1]-1] + 2*img_in[p[0]-1, p[1]] + img_in[(p[0]-1), p[1]+1] - (img_in[(p[0]+1), p[1]-1] + 2*img_in[p[0]+1, p[1]] + img_in[(p[0]+1), p[1]+1])
            result[p] = neighborhood_sum/(6)        

    return result
```
